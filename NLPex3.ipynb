```python
import pandas as pd
import numpy as np
import re
```

```python
data = pd.read_csv('/content/Tweets.csv.zip')
data.info()
data.count()
data['sentiment'].value_counts()
data.head()
```

```python
import pandas as pd # Ensure pandas is imported for pd.notna
import re

pattern = r'\s'
data['char_count'] = data['text'].apply(lambda x: len(re.sub(pattern, '', x if pd.notna(x) else '')))
```

```python
data['char_count']
```

```python
data['word_count'] = data['text'].apply(lambda x: len(str(x).split()))
```
